```python
#phase 4. clustering and segmentation

from sklearn.preprocessing import StandardScaler
X = df.drop('country',axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Data scaled successfully")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,11):
  kmeans = KMeans(
      n_clusters=i,
      random_state=42,
      n_init=10
  )
  kmeans.fit(X_scaled)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Clusters(K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(
    n_clusters = 3,
    random_state=42,
    n_init=10
)
df['cluster'] = kmeans.fit_predict(X_scaled)
print(df['cluster'].value_counts())
```

```python
df[['country','cluster']].head()
```

```python
cluster_summary = df.groupby('cluster').mean(numeric_only=True)
cluster_summary
```

```python
cluster_summary[['income','gdpp','child_mort']]
```

```python
cluster_names={
    0:'Developed',
    1:'Developing',
    2:'Underdeveloped'
}
df['cluster_Name'] = df['cluster'].map(cluster_names)
```

```python
df[['country','cluster','cluster_Name']].head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)
df['PCA1'] = pca_components[:,0]
df['PCA2'] = pca_components[:,1]
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='PCA1',
    y='PCA2',
    hue='cluster_Name',
    palette='Set1',
    s=100
)
plt.title('K-Means Clusters')
plt.show()
```

```python
#dbscan
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(
    eps=1.5,
    min_samples=5
)
df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)
```

```python
num_clusters = len(set(df['DBSCAN_Cluster']))-(
    1 if -1 in df['DBSCAN_Cluster'] else 0
)
print("Number of Clusters:", num_clusters)
```

```python
noise_points = list(df['DBSCAN_Cluster']).count(-1)
print("Number of Noise Points:", noise_points)
```

```python
print(df['DBSCAN_Cluster'].value_counts())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='PCA1',
    y='PCA2',
    hue='DBSCAN_Cluster',
    palette='Set1',
    s=100
)
plt.title('DBSCAN Clusters')
plt.show()
```

```python
kmeans_clusters = df['cluster'].nunique()
dbscan_clusters = len(set(df['DBSCAN_Cluster']))-(
    1 if -1 in df['DBSCAN_Cluster'] else 0
)
print("K-Means Clusters:", kmeans_clusters)
print("DBSCAN Clusters:", dbscan_clusters)
```

```python
from sklearn.metrics import silhouette_score

kmeans_score = silhouette_score(X_scaled, df['cluster'])
dbscan_score = silhouette_score(X_scaled, df['DBSCAN_Cluster'])

print("K-Means Silhouette Score:", kmeans_score)
print("DBSCAN Silhouette Score:", dbscan_score)
```
